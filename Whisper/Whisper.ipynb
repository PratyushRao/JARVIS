```python
import whisper
import kagglehub
import os
import soundfile as sf
import csv
import shutil
from IPython.display import Audio, display
```

Loading jarivs large model and using it to generate transcription

```python
model = whisper.load_model("large")
```

```python
path = kagglehub.dataset_download("fotiemconstant/jarvis-dataset")
print(os.listdir(path))
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
drive_wav_dir = "/content/drive/MyDrive/jarvis_wav"
os.makedirs(drive_wav_dir, exist_ok=True)
```

```python
wav_dir = "/content/jarvis_wav"
os.makedirs(wav_dir, exist_ok=True)

for file in os.listdir(path):
    if file.lower().endswith((".aif", ".aiff")):
        in_path = os.path.join(path, file)
        out_path = os.path.join(wav_dir, file.rsplit(".", 1)[0] + ".wav")

        data, sr = sf.read(in_path)
        sf.write(out_path, data, sr)

        print("Converted:", in_path, "→", out_path)
```

```python
for file_name in os.listdir(wav_dir):
    if file_name.lower().endswith(".wav"):
        src_path = os.path.join(wav_dir, file_name)
        dst_path = os.path.join(drive_wav_dir, file_name)
        shutil.copy(src_path, dst_path)
        print(f"Copied {src_path} to Drive: {dst_path}")
```

```python
transcriptions = {}

for file in os.listdir(wav_dir):
    if file.endswith(".wav"):
        wav_path = os.path.join(wav_dir, file)

        print("Transcribing on GPU:", file)
        result = model.transcribe(wav_path)
        text = result["text"]

        transcriptions[file] = text
        print(" → DONE")
```

```python
csv_path = "/content/jarvis_transcriptions.csv"

with open(csv_path, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)

    writer.writerow(["audio_path", "transcription"])

    for fname, txt in transcriptions.items():
        audio_path = f"/jarvis_wav/{fname}"
        writer.writerow([audio_path, txt.strip()])

print("Saved CSV to:", csv_path)
```
